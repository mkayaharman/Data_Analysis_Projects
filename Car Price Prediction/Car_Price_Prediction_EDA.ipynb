### Some Reminders on Exploratory data analysis (EDA)

Exploratory data analysis (EDA) is an especially important activity in the routine of a data analyst or scientist. It enables an in depth understanding of the dataset, define or discard hypotheses and create predictive models on a solid basis. It uses data manipulation techniques and several statistical tools to describe and understand the relationship between variables and how these can impact business. By means of EDA, we can obtain meaningful insights that can impact analysis under the following questions (If a checklist is good enough for pilots to use every flight, it’s good enough for data scientists to use with every dataset).
1. What question are you trying to solve (or prove wrong)?
2. What kind of data do you have?
3. What’s missing from the data?
4. Where are the outliers?
5. How can you add, change or remove features to get more out of your data?

**``Exploratory data analysis (EDA)``** is often an **iterative brainstorming process** where you pose a question, review the data, and develop further questions to investigate before beginning model development work. The image below shows how the brainstorming phase is connected with that of understanding the variables and how this in turn is connected again with the brainstorming phase.<br>

```python
# Strip and convert "km" column to float

df2["km"] = df2["km"].str.replace(r'\D', '').str.strip()
df2["km"] = df2["km"].replace(r'^\s*$', np.nan, regex=True).astype(float)
```

```python
# Strip and convert "prev_owner" column to float

df2["prev_owner"] = df2["prev_owner"].str.extract('(\d+)').astype(float)
```

```python
# Drop "Previous Owners" column

df2.drop(["Previous Owners"], axis=1, inplace=True)
```

# PART- 2 `( Handling With Missing Vales )`

```python
df2.info()
```

```python
df2["make_model"].value_counts(dropna=False)
```

```python
df2["body_type"].value_counts(dropna=False)
```

```python
# Filling NaN's with "other" for the "body_type" column

df2["body_type"].fillna("Other", inplace=True)
```

```python
df2["vat"].value_counts(dropna=False)
```

```python
# Fill NaN's with "Undescribed" for the "vat" column

df2["vat"].fillna("Undescribed", inplace=True)
```

```python
df2["Age"].value_counts(dropna=False)
```

```python
df2.groupby(["make_model", "Age"])["km"].describe()
```

```python
df2['km'].value_counts(dropna=False)
```

```python
df2["km"] = df2['km'].fillna(df2.groupby(["Age", "make_model"])["km"].transform('mean'))
```

```python
# Fill NaN values of "Age" column with mean value according to group by of "km" and "make_model"

df2["Age"] = df2['Age'].fillna(df2.groupby(["km", "make_model"])["Age"].transform('mean')).apply(np.ceil)
```

```python
# Fill NaN's with "Undescribed" for the "Upholstery" column

df2["Upholstery"].fillna("Undescribed", inplace=True)
```

```python
df2["Fuel"].value_counts(dropna=False)
```

```python
benzine = ["Gasoline", "Gasoline (Particulate Filter)", "Super 95","Super 95 (Particulate Filter)", "Regular", "Super E10 95", "Super E10 95 (Particulate Filter)", "Super Plus 98", "Super Plus E10 98", "Others"]
lpg = ["LPG", "Liquid petroleum gas", "Liquid petroleum gas (LPG)", "CNG", "CNG (Particulate Filter)", "Biogas", "Domestic gas H"]
diesel = ["Diesel", "Diesel (Particulate Filter)"]
def fueltype(x) :
  if x in benzine :
    return "Benzine"
  elif x in lpg :
    return "LPG/CNG"
  elif x in diesel :
    return "Diesel"
  else :
    return x

df2["Fuel"] = df2["Fuel"].apply(fueltype)
```

```python
# Drop columns with more than 25% NaN values

perc = 25.0
min_count = int(((100-perc)/100)*df2.shape[0] + 1)

df2.dropna(axis=1, thresh=min_count, inplace=True)
```

```python
df2.groupby(["Fuel"])["Emission Class"].agg(pd.Series.mode)
```

```python
# Fill NaN values of "Emission Class" column with mean value according to group by of "Fuel"

df2['Emission Class'] = df2['Emission Class'].fillna(df2["Emission Class"].mode()[0])
```

```python
df2.groupby(["make_model", "Age"])["CO2_Emission_g"].mean()
```

```python
df2["CO2_Emission_g"] = df2['CO2_Emission_g'].fillna(df2.groupby(["Age", "make_model"])["CO2_Emission_g"].transform('mean'))
```

```python
df2['CO2_Emission_g'].value_counts(dropna=False)
```

```python
df2.groupby(["make_model", "Age"])["Consumption"].mean()
```

```python
df2["Consumption"] = df2['Consumption'].fillna(df2.groupby(["Age", "make_model"])["Consumption"].transform('mean'))
```

```python
df2.groupby(["make_model"])["Comfort_Convenience"].agg(pd.Series.mode)
```

```python
df2["Comfort_Convenience"] = df2["Comfort_Convenience"].fillna(df2["Comfort_Convenience"].mode()[0])
```

```python
df2["Entertainment_Media"] = df2["Entertainment_Media"].fillna(df2["Entertainment_Media"].mode()[0])
```

```python
df2.groupby(["make_model"])["Extras"].agg(pd.Series.mode)
```

```python
df2["Extras"] = df2["Extras"].fillna(df2["Extras"].mode()[0])
```

```python
df2["Safety_Security"] = df2["Safety_Security"].fillna(df2["Extras"].mode()[0])
```

```python
def drop_outliers(df, field_name) :
  iqr = 1.5 * (np.percentile(df[field_name], 75) - np.percentile(df[field_name], 25))
  df.drop(df[df[field_name] > (iqr + np.percentile(df[field_name], 75))].index, inplace = True)
  df.drop(df[df[field_name] < (iqr + np.percentile(df[field_name], 25))].index, inplace = True)
```
